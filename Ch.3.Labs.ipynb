```python
# 3.3.1. Loads up the seeds dataset, located at Module3/Datasets/wheat.data' into a dataframe. You should be very good at doing this by now.
import pandas as pd
seeds = pd.read_csv('Datasets/wheat.data')
seeds.head()
```

```python
# import 3D library
from mpl_toolkits.mplot3d import Axes3D 
```

```python
# 3.3.2. Graph a 3D scatter plot using the area, perimeter, and asymmetry features. 
#  Be sure to label your axes, and use the optional display parameter c='red'.
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d') 
ax.set_xlabel('Area') 
ax.set_ylabel('Perimeter') 
ax.set_zlabel('Asymmetry') 
ax.scatter(seeds['area'], seeds['perimeter'], seeds['asymmetry'], 
    c='r', marker='o') 
plt.show() 
```

```python
# 3.3.3. Graph a 3D scatter plot using the width, groove, and length features. 
# Be sure to label your axes, and use the optional display parameter c='green'.
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d') 
ax.set_xlabel('Width') 
ax.set_ylabel('Groove') 
ax.set_zlabel('Length') 
ax.scatter(seeds['width'], seeds['groove'], seeds['length'], 
    c='g', marker='o') 
plt.show() 
```

```python
# 3.6.1. Loads up the seeds dataset, located at Module3/Datasets/wheat.data into a dataframe
import pandas as pd
seeds6 = pd.read_csv('Datasets/wheat.data')
seeds6.head()
```

```python
# 3.6.2. Drop the id column from the dataframe
seeds6 = seeds6.drop(labels=['id'], axis=1)
seeds6.head()
```

```python
# 3.6.3. Compute the correlation matrix of your dataframe
seeds6.corr()
```

```python
# 3.6.4. Graph the correlation matrix using imshow or matshow (either or)
plt.figure(figsize=(10,10))

plt.imshow(seeds6.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(seeds6.columns))]
plt.xticks(tick_marks, seeds6.columns, rotation='vertical')
plt.yticks(tick_marks, seeds6.columns)

plt.show()
```
